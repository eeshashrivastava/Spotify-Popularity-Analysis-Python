```python
import pandas as pd

df = pd.read_csv("../data_processed/spotify_clean.csv")
df.head()
```

```python
df["popularity_bucket"].value_counts(normalize=True)
```

```python
summary = df.groupby("popularity_bucket")[
    ["danceability", "energy", "valence", "acousticness", "tempo"]
].mean()

summary
```

The analysis shows clear and consistent differences in audio characteristics across popularity levels, suggesting that musical features play a meaningful role in track success.

Highly popular tracks tend to be more energetic and danceable, indicating that upbeat, high-energy songs are more likely to resonate with a broad audience. These tracks also exhibit lower acousticness, suggesting a more produced and polished sound that aligns with mainstream listening preferences.

Moderately popular tracks fall between high and low popularity groups across most features, indicating that popularity exists on a spectrum rather than as a binary outcome. This suggests that incremental changes in audio features may contribute to increases in popularity.

Low popularity tracks are generally more acoustic, slightly slower, and lower in energy. This implies that while these songs may appeal to niche audiences, they are less likely to achieve mass appeal without stronger energetic or rhythmic elements.

Overall, the results suggest that popularity is driven by a combination of features rather than any single metric. Energy, danceability, and tempo together appear to be more predictive than tempo alone. From a business perspective, this highlights how artists, producers, and labels may optimize multiple audio attributes simultaneously when aiming for broader reach and commercial success.
