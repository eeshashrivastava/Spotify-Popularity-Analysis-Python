```python
import pandas as pd

df = pd.read_csv("./data:raw:/dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(columns=["Unnamed: 0"])
df.shape
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
keep_cols = [
    "track_id",
    "track_name",
    "artists",
    "album_name",
    "track_genre",
    "popularity",
    "duration_ms",
    "explicit",
    "danceability",
    "energy",
    "loudness",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence",
    "tempo"
]

df = df[keep_cols]
df.head()
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
def popularity_bucket(x):
    if x < 40:
        return "Low"
    elif x < 70:
        return "Medium"
    else:
        return "High"

df["popularity_bucket"] = df["popularity"].apply(popularity_bucket)
df["popularity_bucket"].value_counts()
```

```python
def tempo_bucket(t):
    if t < 90:
        return "Slow"
    elif t < 120:
        return "Medium"
    else:
        return "Fast"

df["tempo_bucket"] = df["tempo"].apply(tempo_bucket)
df["tempo_bucket"].value_counts()
```

```python
df.to_csv("data:processed:/spotify_clean.csv", index=False)
```
